## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st

# Study data files
mouse_metadata_path = "Resources/Mouse_metadata.csv"
study_results_path = "Resources/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
```

```python
mouse_metadata.head()
```

```python
study_results.head()
```

```python
# Combine the data into a single dataset
# Display the data table for preview

combined_data = pd.merge(mouse_metadata, study_results, how="outer", on="Mouse ID")
combined_data.head()
```

```python
# Checking the number of mice.

mice_number = len(combined_data["Mouse ID"].unique())
print(f"There are a total of {mice_number} mice in the study.")
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
# Optional: Get all the data for the duplicate mouse ID. 

dup_mouse = combined_data.loc[combined_data.duplicated(subset=["Mouse ID", "Timepoint"]),"Mouse ID"].unique()
combined_data[combined_data["Mouse ID"]=="g989"]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_DF = combined_data[combined_data["Mouse ID"].isin(dup_mouse)==False]
clean_DF
```

```python
# Checking the number of mice in the clean DataFrame

cleanmouse = len(clean_DF["Mouse ID"].unique())
print(f"There are {cleanmouse} in the study.")
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Using the aggregation method, produce the same summary statistics in a single line
sumdf = clean_DF.loc[:, ["Mouse ID", "Drug Regimen", "Tumor Volume (mm3)"]]
sumdf

mean = sumdf.groupby(["Drug Regimen"]).mean()["Tumor Volume (mm3)"]
median = sumdf.groupby(["Drug Regimen"]).median()["Tumor Volume (mm3)"]
variance = sumdf.groupby(["Drug Regimen"]).var()["Tumor Volume (mm3)"]
standard_deviation = sumdf.groupby(["Drug Regimen"]).std()["Tumor Volume (mm3)"]
SEM = sumdf.groupby(["Drug Regimen"]).sem()["Tumor Volume (mm3)"]

stats = pd.DataFrame({
    "Mean TV (mm3)":mean,
    "Median TV (mm3)":median,
    "Variance TV (mm3)": variance,
    "STDEV TV (mm3)": standard_deviation,
    "SEM TV (mm3)": SEM
})
stats_DF = stats.round(2)
stats_DF
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

mouse_weight = Capomulin_df.groupby(Capomulin_df["Mouse ID"])["Weight (g)"].mean()
tumor_volume = Capomulin_df.groupby(Capomulin_df["Mouse ID"])["Tumor Volume (mm3)"].mean()

slope, int, r, p, std_err = st.linregress(mouse_weight, tumor_volume)
      
fit = slope * mouse_weight + int

plt.scatter(mouse_weight,tumor_volume)
plt.xlabel("Weight of Mouse")
plt.ylabel("Tumor Volume")
plt.plot(mouse_weight,fit,"--")
plt.xticks(mouse_weight, rotation=90)
plt.show()

corr = round(st.pearsonr(mouse_weight,tumor_volume)[0],2)
print(f'The correlation between weight and tumor value is {corr}')
```
